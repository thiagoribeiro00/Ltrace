```python
import h5py
import numpy as np
import cv2
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import matplotlib.pyplot as plt
import torchvision.utils as vutils
import os

# Carregar o arquivo HDF5
file_path = '/content/drive/MyDrive/train_images.h5'
with h5py.File(file_path, 'r') as f:
    dataset_name = list(f.keys())[0]
    images = np.array(f[dataset_name])

# Redimensionar e converter as imagens para escala de cinza
images_resized = []
for img in images:
    resized_img = cv2.resize(img, (64, 64))  # Redimensionar para 64x64
    if resized_img.ndim == 3:  # Se a imagem tiver 3 canais
        resized_img = cv2.cvtColor(resized_img, cv2.COLOR_RGB2GRAY)  # Converter para escala de cinza
    # Normalizar a imagem para o intervalo [0, 1]
    resized_img = resized_img / 255.0  # Normalização para [0, 1]
    images_resized.append(resized_img)

# Converter as imagens para tensor e adicionar dimensão de canal
images_tensor = torch.tensor(np.array(images_resized), dtype=torch.float32).unsqueeze(1)

# Criar um DataLoader
batch_size = 64
dataset = TensorDataset(images_tensor)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# Detectar dispositivo (GPU ou CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Função de inicialização de pesos (Xavier)
def weights_init(m):
    if isinstance(m, nn.Conv2d) or isinstance(m, nn.ConvTranspose2d):
        nn.init.xavier_uniform_(m.weight)
        if m.bias is not None:
            nn.init.zeros_(m.bias)

class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(100, 128 * 16 * 16),  # Aumentar a dimensão
            nn.ReLU(),
            nn.BatchNorm1d(128 * 16 * 16),
            nn.Unflatten(1, (128, 16, 16)),  # Reformatar para 128 canais de 16x16
            nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1),  # De 16x16 para 32x32
            nn.ReLU(),
            nn.ConvTranspose2d(64, 1, kernel_size=4, stride=2, padding=1),  # De 32x32 para 64x64
            nn.Sigmoid()  # Saída no intervalo [0, 1]
        )

    def forward(self, z):
        return self.model(z)

class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.model = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, stride=2, padding=1),
            nn.LeakyReLU(0.2),
            nn.Conv2d(16, 32, kernel_size=3, stride=2, padding=1),
            nn.LeakyReLU(0.2),
            nn.Flatten(),
            nn.Linear(32 * 16 * 16, 1)
        )

    def forward(self, x):
        return torch.sigmoid(self.model(x))

# Função para calcular o gradient penalty
def gradient_penalty(discriminator, real_data, fake_data, lambda_gp=10):
    batch_size = real_data.size(0)
    epsilon = torch.rand(batch_size, 1, 1, 1, device=real_data.device)
    interpolated = epsilon * real_data + (1 - epsilon) * fake_data
    interpolated.requires_grad_(True)

    d_interpolated = discriminator(interpolated)
    gradients = torch.autograd.grad(outputs=d_interpolated,
                                    inputs=interpolated,
                                    grad_outputs=torch.ones_like(d_interpolated),
                                    create_graph=True,
                                    only_inputs=True)[0]

    grad_norm = gradients.view(batch_size, -1).norm(2, dim=1).mean()
    gp = (grad_norm - 1) ** 2
    return lambda_gp * gp

# Configurações
latent_dim = 100
num_epochs = 200  # Aumentar o número de épocas
learning_rate = 0.0001  # Ajustar a taxa de aprendizado
beta1 = 0.5

# Inicializando os modelos, otimizadores e funções de perda
generator = Generator().to(device)
discriminator = Discriminator().to(device)
generator.apply(weights_init)
discriminator.apply(weights_init)

optimizer_G = optim.Adam(generator.parameters(), lr=learning_rate, betas=(beta1, 0.999))
optimizer_D = optim.Adam(discriminator.parameters(), lr=learning_rate, betas=(beta1, 0.999))

criterion = nn.BCELoss()
lambda_gp = 10  # Coeficiente do gradient penalty

# Função para salvar imagens geradas
def save_generated_images(epoch, fixed_noise):
    # Criar o diretório se não existir
    os.makedirs('images', exist_ok=True)

    with torch.no_grad():
        fake_images = generator(fixed_noise).detach().cpu()
        vutils.save_image(fake_images, f'images/epoch_{epoch}.png', normalize=True)

# Inicializando o ruído fixo para visualização
fixed_noise = torch.randn(64, latent_dim, device=device)  # Alterado para 64 para corresponder ao número de imagens a serem exibidas

# Loop de treinamento
d_losses = []
g_losses = []

for epoch in range(num_epochs):
    for i, (real_images,) in enumerate(dataloader):
        real_images = real_images.to(device)
        real_labels = torch.ones(real_images.size(0), 1, device=device)
        fake_labels = torch.zeros(real_images.size(0), 1, device=device)

        # Atualizar o discriminador
        optimizer_D.zero_grad()
        outputs = discriminator(real_images)
        d_loss_real = criterion(outputs, real_labels)

        z = torch.randn(real_images.size(0), latent_dim, device=device)
        fake_images = generator(z)
        outputs = discriminator(fake_images.detach())
        d_loss_fake = criterion(outputs, fake_labels)

        gp = gradient_penalty(discriminator, real_images, fake_images)
        d_loss = d_loss_real + d_loss_fake + gp
        d_loss.backward(retain_graph=True)  # Adicione retain_graph=True aqui
        optimizer_D.step()

        # Atualizar o gerador
        optimizer_G.zero_grad()
        outputs = discriminator(fake_images)
        g_loss = criterion(outputs, real_labels)
        g_loss.backward()  # Aqui, você não precisa de retain_graph=True
        optimizer_G.step()

    # Armazenar e exibir perdas
    d_losses.append(d_loss.item())
    g_losses.append(g_loss.item())

    print(f'Época [{epoch + 1}/{num_epochs}], Perda D: {d_loss.item():.4f}, Perda G: {g_loss.item():.4f}')
    if (epoch + 1) % 10 == 0:
        save_generated_images(epoch + 1, fixed_noise)
        plt.figure(figsize=(10, 5))
        plt.plot(d_losses, label='Perda Discriminador')
        plt.plot(g_losses, label='Perda Gerador')
        plt.xlabel('Época')
        plt.ylabel('Perda')
        plt.legend()
        plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Função para exibir imagens geradas
def show_generated_images(epoch, fixed_noise):
    with torch.no_grad():
        fake_images = generator(fixed_noise).detach().cpu()

    fig, ax = plt.subplots(figsize=(8, 8))
    ax.set_xticks([]); ax.set_yticks([])
    ax.imshow(np.transpose(vutils.make_grid(fake_images[:64], padding=2, normalize=True), (1, 2, 0)))
    plt.title(f"Epoch {epoch}")
    plt.show()

# Inicializando o ruído fixo para visualização
fixed_noise = torch.randn(64, latent_dim, device=device)

# Exibir as imagens geradas após o treinamento
show_generated_images(num_epochs, fixed_noise)
```
